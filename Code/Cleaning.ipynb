```python
import pandas as pd
```

```python
files = ['aqarmap_data_2_7.csv','aqarmap_data_8_15.csv','aqarmap_data_16_22.csv']

dataframes = []

for file in files :
    df = pd.read_csv(file)
    dataframes.append(df)
    
DF = pd.concat(dataframes,ignore_index=True)

print(len(DF))

DF_cleaned = DF.drop_duplicates(subset=['URL'])

print(len(DF_cleaned))
```
